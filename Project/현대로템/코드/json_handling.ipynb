```python
import json
import numpy as np
from statistics import mean, median, mode, stdev

json_data = {str(i): {"hands": random.randint(0, 1)} for i in range(1, 101)}

def calculate_statistics():
    num = []
    for i in range(100):
        key = str(i + 1)
        num.append(json_data[key]['hands'])
    
    # 기술통계량 계산
    avg = mean(num)              # 평균
    med = median(num)            # 중앙값
    mod = mode(num)              # 최빈값
    std = stdev(num) if len(num) > 1 else 0  # 표준편차 (표본 크기 1 이상일 때만 계산)
    min_val = min(num)           # 최소값
    max_val = max(num)           # 최대값
    count_0 = num.count(0)       # 0의 개수
    count_1 = num.count(1)       # 1의 개수
    
    np_array = np.array(num)
    q1 = np.percentile(np_array, 25)  # 1사분위수
    q3 = np.percentile(np_array, 75)  # 3사분위수
    
    # 결과 출력
    print(f"평균: {avg:.3f}")
    print(f"중앙값: {med}")
    print(f"최빈값: {mod}")
    print(f"표준편차: {std:.3f}")
    print(f"최소값: {min_val}")
    print(f"최대값: {max_val}")
    print(f"0의 개수: {count_0}")
    print(f"1의 개수: {count_1}")
    print(f"1사분위수(Q1): {q1}")
    print(f"3사분위수(Q3): {q3}")

def main():
    print(">>> calculate_statistics()")
    calculate_statistics()

if __name__ == "__main__":
    main()
```

```python
import json
import numpy as np
from statistics import mode, StatisticsError
import matplotlib.pyplot as plt

def calculate_statistics(num):
    # 기술통계량 계산
    avg = np.mean(num)              # 평균
    med = np.median(num)            # 중앙값
    try:
        mod = mode(num)             # 최빈값
    except StatisticsError:
        mod = "없음 (동일 빈도)"    # 최빈값이 없을 경우
    std = np.std(num) if len(num) > 1 else 0  # 표준편차 (표본 크기 1 이상일 때만 계산)
    min_val = np.min(num)           # 최소값
    max_val = np.max(num)           # 최대값
    count_0 = num.count(0)          # 0의 개수
    count_1 = num.count(1)          # 1의 개수
    
    np_array = np.array(num)
    q1 = np.percentile(np_array, 25)  # 1사분위수
    q3 = np.percentile(np_array, 75)  # 3사분위수
    
    # 통계량 딕셔너리로 반환 (시각화를 위해)
    stats = {
        "평균": avg,
        "중앙값": med,
        "최빈값": mod,
        "표준편차": std,
        "최소값": min_val,
        "최대값": max_val,
        "0의 개수": count_0,
        "1의 개수": count_1,
        "1사분위수(Q1)": q1,
        "3사분위수(Q3)": q3
    }
    return stats, num

# JSON 데이터 로드
with open("C:/Users/acorn/Desktop/code/video_detection_results.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 모든 프레임의 "class" 값 수집
last_key = list(data)[-1]["frame_id"] + 1  # 최대 frame_id + 1
classValue = []
for frame in data:
    for detection in frame["detections"]:
        classValue.append(detection["class"])

# 통계량 계산
stats, data_array = calculate_statistics(classValue)

# 시각화
plt.figure(figsize=(12, 6))

# 1. 히스토그램
plt.subplot(1, 2, 1)
plt.hist(data_array, bins=range(int(min(data_array)), int(max(data_array)) + 2), edgecolor='black')
plt.title('Class 값 분포 (히스토그램)')
plt.xlabel('Class 값')
plt.ylabel('빈도')

# 2. 박스 플롯
plt.subplot(1, 2, 2)
plt.boxplot(data_array)
plt.title('Class 값 분포 (박스 플롯)')
plt.ylabel('Class 값')

# 통계량 텍스트 추가
stats_text = "\n".join([f"{k}: {v:.3f}" if isinstance(v, (int, float)) else f"{k}: {v}" for k, v in stats.items()])
plt.figtext(0.91, 0.5, stats_text, ha="left", va="center", fontsize=10, bbox={"facecolor": "white", "alpha": 0.5, "pad": 5})

# 레이아웃 조정 및 출력
plt.tight_layout()
plt.show()
```

```python
import json
import numpy as np
from statistics import mode, StatisticsError
import matplotlib.pyplot as plt
from collections import Counter

plt.rcParams['font.family'] = 'Malgun Gothic'  
plt.rcParams['axes.unicode_minus'] = False  

def calculate_statistics(num):
    counter = Counter(num)  
    total_count = len(num) 
    try:
        mod = mode(num)     
    except StatisticsError:
        mod = "없음 (동일 빈도)"
    
    # 클래스별 빈도수와 비율
    class_counts = dict(counter)
    class_ratios = {k: v / total_count * 100 for k, v in class_counts.items()}  # 비율 (%)

    stats = {
        "최빈값": mod,
        "클래스별 빈도수": class_counts,
        "클래스별 비율(%)": class_ratios,
        "총 데이터 개수": total_count
    }
    return stats
    
with open("C:/Users/acorn/Desktop/code/video_detection_results.json", "r", encoding="utf-8") as f:
    data = json.load(f)

classValue = []
for frame in data:
    for detection in frame["detections"]:
        classValue.append(detection["class"])

stats = calculate_statistics(classValue)

# 시각화
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
classes = list(stats["클래스별 빈도수"].keys())
counts = list(stats["클래스별 빈도수"].values())
plt.bar(classes, counts, color='skyblue', edgecolor='black')
plt.title('클래스별 빈도수 (막대 그래프)')
plt.xlabel('Class 값')
plt.ylabel('빈도')
plt.xticks(classes)

plt.subplot(1, 2, 2)
ratios = list(stats["클래스별 비율(%)"].values())
labels = [f"Class {k}" for k in stats["클래스별 비율(%)"].keys()]
plt.pie(ratios, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel1.colors)
plt.title('클래스별 비율 (파이 차트)')
plt.axis('equal')  # 원형 유지

stats_text = (
    f"최빈값: {stats['최빈값']}\n"
    f"총 데이터 개수: {stats['총 데이터 개수']}\n\n"
    "클래스별 빈도수:\n" + "\n".join([f"Class {k}: {v}" for k, v in stats["클래스별 빈도수"].items()]) + "\n\n"
    "클래스별 비율(%):\n" + "\n".join([f"Class {k}: {v:.1f}%" for k, v in stats["클래스별 비율(%)"].items()])
)
plt.figtext(0.91, 0.5, stats_text, ha="left", va="center", fontsize=10, bbox={"facecolor": "white", "alpha": 0.5, "pad": 5})

plt.tight_layout()
plt.show()
```
